    b. Add the product name to the "customer_orders" set.

3. Instead of updating the inventory by subtracting 1 from the quantity of each product, only do it for  the products that were ordered (those in "customer_orders").

```python
products = ['t-shirt', 'mug', 'hat', 'book', 'keychain']

inventory = {}
for item in products:
    inventory[item] = int(input(f'Insert {item} quantity: '))

customers_orders = set()
    
order_products = input('Enter the product that the customer wants to order: ').lower().strip()

while order_products not in products:
    print('Invalid product! It needs to be one of the following products: t-shirt mug hat book keychain.')
    order_products = input('Enter the product that the customer wants to order: ').lower().strip()

customers_orders.add(order_products)
another_product = input('Do you want to add another product? yes/no: ').lower().strip()

while another_product == "yes":
    order_products = input('Enter the product that the customer wants to order: ').lower().strip()

    while order_products not in products:
        print('Invalid product! It needs to be one of the following products: t-shirt mug hat book keychain.')
        order_products = input('Enter the product that the customer wants to order: ').lower().strip()
    
    customers_orders.add(order_products)
    another_product = input('Do you want to add another product? yes/no: ').lower().strip()


for item in customers_orders:
    inventory[item] -= 1

for item, quantity in inventory.items():
    print(f'{item} - {quantity}')
    
```
